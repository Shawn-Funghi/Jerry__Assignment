```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([2,4,6,8,10,12,14,16])
y = np.array([10,50,70,75,40,10,30,35])
A = np.vstack([x**3,x**2,x, np.ones(len(x))]).T
result =  np.linalg.lstsq(A,y,rcond=None)
a,b,c,d =result[0]
```

```python
ran = np.linspace(2,16,100)
_ = plt.plot(x, y, 'o', label='Original data', markersize=10)
_ = plt.plot(ran, a*ran**3+b*ran**2+c*ran + d, 'r', label='Fitted line')
_ = plt.legend()
plt.show()
```
